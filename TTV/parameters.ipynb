```python
save = True
T_s = Parameter(r'$T_{eff}$',[4425,70],'Piaulet et al. 2021')
Metallicity_s = Parameter(r'$[Fe/H]_{s}$',[0.020,0.090],'Piaulet et al. 2021')
logg_s = Parameter(r'$log(g)_{s}$',[4.633,0.012],'Piaulet et al. 2021')
rp = Parameter(r'$R_p/R_{s}$',[0.1446,0.0002],'Kokori et al. 2023')
period = Parameter(r'$Period$',[5.72148926,0.00000085],'Kokori et al. 2023')
a_R = Parameter(r'$a/R_s$',[17.96,0.05],'Kokori et al. 2023')
inc = Parameter(r'$Inclination$',[89.56,0.08],'Kokori et al. 2023')
b = Parameter(r'$b$',np.cos(inc.value[0]*np.pi/180)*a_R.value[0],'Kokori et al. 2023') # b=cos(i)*a/Rs
mass_p = Parameter(r'$M_p$',[0.0960,0.0050],'Piaulet et al. 2021')
radius_p = Parameter(r'$R_p$',[0.948,0.030],'Dai & Winn 2017')
T_p = Parameter(r'$T_{eq}$',[736,17],'Močnik et al. 2017')
mass_s = Parameter(r'$M_s$',[0.683,0.017,0.016],'Piaulet et al. 2021')
radius_s = Parameter(r'$R_s$',[0.67,0.02],'Piaulet et al. 2021')
eccentricity = Parameter(r'$Eccentricity$',[0.06,0.04],'Piaulet et al. 2021')
omega = Parameter(r'$\omega$',[40,40,-60],'Piaulet et al. 2021')
parameters = [T_s,Metallicity_s,logg_s,rp,period,a_R,inc,b,mass_p,radius_p,T_p,mass_s,radius_s,eccentricity,omega]
WASP_107b = Planet('WASP-107b',parameters)
if save: path='wasp-107b/'
else: path=None
WASP_107b.printParameters(path=path)
```
